```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Pandas_Tutorial_6_Handle_Missing_Data_CSV.csv",parse_dates=['day'])
df
```

```python
df.set_index("day",inplace=True)
df
```

```python
new_df = df.replace([-99999, -88888],np.NaN)
new_df
```

```python
new_df = df.replace({
    'temperature':-99999,
    'windspeed':[-99999,-88888],
    'event':'0'
},np.NaN)
new_df
```

```python
new_df = df.replace({
    "-99999":np.NaN,
    "-88888":np.NaN,
    'No Event':'Sunny'
})
new_df
```

```python
new_df = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
new_df
```

```python
df = pd.DataFrame({
    'score':['exceptional','average','good','poor','average','exceptional'],
    'student':['rob','maya','parthiv','tom','julian','erica']
})
df
```
